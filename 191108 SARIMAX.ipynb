```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import plotly.express as px
import plotly.graph_objects as go

from statsmodels.tsa.statespace.sarimax import SARIMAX

%matplotlib inline
```

# 데이터 불러오기 (count1718)

```python
count1718 = pd.read_csv('../data/2017_2018_운영시간내_실이용객_시각별_주차장_카운트.csv')
count1718
```

# Seaborn lineplot for SRIMAX

```python
# plt.figure(figsize=(20,10))
# sns.lineplot('datetime', 'count', data=count1718)
```

```python
# 2018 1월 예측
# 23시간 * 31일
x_train = count1718.iloc[:-713, 1]
x_test = count1718.iloc[-713:, 1]
```

```python
res = SARIMAX(
    x_train, 
    # p, d, q
    order=(1, 1, 1), 
    # P, D, Q, m
    seasonal_order=(1, 1, 1, 1)
).fit()
```

```python
res.summary()
```

### 예측

```python
n = len(x_train)
m = len(x_test)
forecast = res.predict(n + 1, n+m)
```

```python
sns.lineplot('datetime', 'count', data=count1718, color='lightgrey')
sns.lineplot(
    count1718.iloc[-713:, 0],
    forecast,
    color='red'
)
```
