```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
wc_17 = pd.read_csv('./data/[회귀]요일_날씨_입출구_2017.csv')
wc_18 = pd.read_csv('./data/[회귀]요일_날씨_입출구_2018.csv')
wc_19 = pd.read_csv('./data/[회귀]요일_날씨_입출구_2019.csv')

holi = pd.read_csv('./data/[회귀]all_공휴일.csv')
```

```python
wc_17['datetime'] = pd.to_datetime(wc_17['datetime'])
wc_18['datetime'] = pd.to_datetime(wc_18['datetime'])
wc_19['datetime'] = pd.to_datetime(wc_19['datetime'])
```

# 공휴일 날짜데이터 형식으로 변환

```python
holi['locdate'] = holi['locdate'].map(lambda d: str(d))
```

```python
holi['locdate'] = pd.to_datetime(holi['locdate'].map(lambda d: f'{d[:4]}-{d[4:6]}-{d[6:]}'))
```

```python
holi['is_holiday'] = 1

holi['year'] = holi['locdate'].dt.year
holi['month'] = holi['locdate'].dt.month
holi['date'] = holi['locdate'].dt.day
```

```python
holi
```

```python
wch_17 = wc_17.merge(holi, how='left').fillna(0)
wch_17
```

```python
wch_18 = wc_18.merge(holi, how='left').fillna(0)
```

```python
wch_19 = wc_19.merge(holi, how='left').fillna(0)
wch_19
```

```python
wch_17.to_csv('./data/[회귀]최종_전처리_완료_요일_날씨_공휴일_2017.csv', index=False)
wch_18.to_csv('./data/[회귀]최종_전처리_완료_요일_날씨_공휴일_2018.csv', index=False)
wch_19.to_csv('./data/[회귀]최종_전처리_완료_요일_날씨_공휴일_2019.csv', index=False)
```
