```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
alldf_15_780 = pd.read_csv('./data/전체_주차시간기준_전처리.csv')
```

```python
alldf_15_780
```

```python
alldf_15_780['진입시각'] = pd.to_datetime(alldf_15_780['진입시각'])
```

```python
alldf_15_780['진출시각'] = pd.to_datetime(alldf_15_780['진출시각'])
```

```python
alldf_15_780.dtypes
```

## 날짜 변경될때까지 주차되어있는 친구들 제거

```python
alldf_15_780['진입_년_월_일'] = alldf_15_780['진입시각'].apply(lambda t: t.strftime('%Y-%m-%d'))
```

```python
alldf_15_780['진출_년_월_일'] = alldf_15_780['진출시각'].apply(lambda t: t.strftime('%Y-%m-%d'))
```

```python
samedate_alldf = alldf_15_780[alldf_15_780['진입_년_월_일'] == alldf_15_780['진출_년_월_일']]
samedate_alldf
```

## 영업시간내에 진입 진출한 경우만 남기기

```python
samedate_alldf['in_hour'] = samedate_alldf['진입시각'].dt.hour
```

```python
samedate_alldf['out_hour'] = samedate_alldf['진출시각'].dt.hour
```

```python
samedate_alldf[(samedate_alldf['in_hour'] >= 9) & (samedate_alldf['out_hour'] <= 22)]
```

```python
samedate_alldf.to_csv('./data/전체_운영시간내_실이용객_')
```
